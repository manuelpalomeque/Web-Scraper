```python
# Importar librerias necesarias:
import pandas as pd
```

```python
# Leer los archivos .csv:

# Hiper Libertad:
df_hiper_libertad_1 = pd.read_csv('/content/drive/MyDrive/Web Scrapping - Supermercados/Hiper_Libertad/relevamiento_precios_hiper_libertad.csv')
df_hiper_libertad_2 = pd.read_csv('/content/drive/MyDrive/Web Scrapping - Supermercados/Hiper_Libertad/relevamiento_precios_hiper_libertad1.csv')
df_hiper_libertad_3 = pd.read_csv('/content/drive/MyDrive/Web Scrapping - Supermercados/Hiper_Libertad/relevamiento_precios_hiper_libertad2.csv')
df_hiper_libertad_4 = pd.read_csv('/content/drive/MyDrive/Web Scrapping - Supermercados/Hiper_Libertad/relevamiento_precios_hiper_libertad3.csv')

# Disco:
df_disco_1 = pd.read_excel('/content/drive/MyDrive/Web Scrapping - Supermercados/Disco/relevamiento_precios_disco.xlsx')
df_disco_2 = pd.read_csv('/content/drive/MyDrive/Web Scrapping - Supermercados/Disco/relevamiento_precios_disco1.csv')
df_disco_3 = pd.read_csv('/content/drive/MyDrive/Web Scrapping - Supermercados/Disco/relevamiento_precios_disco2.csv')
df_disco_4 = pd.read_csv('/content/drive/MyDrive/Web Scrapping - Supermercados/Disco/relevamiento_precios_disco3.csv')
```

```python
# Unirlos por super mercado:
df_hiper_libertad = pd.concat([df_hiper_libertad_1, df_hiper_libertad_2, df_hiper_libertad_3, df_hiper_libertad_4], ignore_index=True)
df_disco = pd.concat([df_disco_1, df_disco_2, df_disco_3, df_disco_4], ignore_index=True)

# Agregar una columna "supermercado" con el valor correspondiente a cada uno:
df_hiper_libertad['Supermercado'] = 'Hiper Libertad'
df_disco['Supermercado'] = 'Disco'
```

```python
# Dataframe final y unificado:
df_unificado = pd.concat([df_disco, df_hiper_libertad], ignore_index=True)
```

```python
# Convertir la columna fecha en formato de fecha
df_unificado['fecha'] = pd.to_datetime(df_unificado['fecha'], format = '%d-%m-%Y')
```

```python
# Excluir registros de precios "No disponibles"
df_unificado = df_unificado[df_unificado['Precio'] != "No disponible"]
```

```python
# Convertir la columna precio en formato de float:

# hay registros que tienen los precios con coma y un espacio en el valor decimal, lo cual
# impide transformar dicha columna directamente a float, por lo que tengo que realizar
# transformaciones sobre la columna para poder continuar.

# Reemplazar la coma por punto en aquellos registros que contengan "","" y eliminar espacios
for index, value in df_unificado['Precio'].items():
  if ',' in str(value):
    df_unificado.at[index, 'Precio'] = str(value).replace(',', '.').replace(' ', '')

# Convertir a float
df_unificado['Precio'] = df_unificado['Precio'].astype(float)
```

```python
# seleccionar columnas necesarias para el analisis
df_unificado = df_unificado[['fecha', 'Nombre Producto', 'Precio', 'Supermercado']]
```
