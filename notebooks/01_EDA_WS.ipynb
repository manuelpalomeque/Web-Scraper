## 3. Limpieza de datos

```python
# Convertir la columna fecha en formato de fecha
df_unificado['fecha'] = pd.to_datetime(df_unificado['fecha'], format = '%d-%m-%Y')
```

```python
# Excluir registros de precios "No disponibles"
df_unificado = df_unificado[df_unificado['Precio'] != "No disponible"]
```

```python
# Convertir la columna precio en formato de float:

# hay registros que tienen los precios con coma y un espacio en el valor decimal, lo cual
# impide transformar dicha columna directamente a float, por lo que tengo que realizar
# transformaciones sobre la columna para poder continuar.

# Reemplazar la coma por punto en aquellos registros que contengan "","" y eliminar espacios
for index, value in df_unificado['Precio'].items():
  if ',' in str(value):
    df_unificado.at[index, 'Precio'] = str(value).replace(',', '.').replace(' ', '')

# Convertir a float
df_unificado['Precio'] = df_unificado['Precio'].astype(float)
```

```python
# seleccionar columnas necesarias para el analisis
df_unificado = df_unificado[['fecha', 'Nombre Producto', 'Precio', 'Supermercado']]
```

```python
# Previsualizar primeros registros
df_unificado.head()
```

```python
# informacion general del dataframe
df_unificado.info()
```

```python
df_unificado.describe()
```

```python
# Multiplicar por dos el valor del precio de la batata en el supermercado
# hiper libertad, ya que  su precio es por medio kilogramos y el precio de disco
# es por 1 kilogramo

for i in range(len(df_unificado)):
  if df_unificado.iloc[i]['Supermercado'] == 'Hiper Libertad':
    if 'batata' in df_unificado.iloc[i]['Nombre Producto'].lower():
      df_unificado.iloc[i, df_unificado.columns.get_loc('Precio')] = (
                df_unificado.iloc[i]['Precio'] * 2
            )
```

```python
# Multiplicar por dos el valor del precio de la cebolla en el supermercado
# hiper libertad, ya que  su precio es por medio kilogramos y el precio de disco
# es por 1 kilogramo

for i in range(len(df_unificado)):
  if df_unificado.iloc[i]['Supermercado'] == 'Hiper Libertad':
    if 'cebolla' in df_unificado.iloc[i]['Nombre Producto'].lower():
      df_unificado.iloc[i, df_unificado.columns.get_loc('Precio')] = (
                df_unificado.iloc[i]['Precio'] * 2
            )
```

```python
# Multiplicar por dos el valor del precio del pollo en el supermercado
# disco, ya que  su precio es por kilogramos y el precio de hiper libertad
# es por 2 kilogramo

for i in range(len(df_unificado)):
  if df_unificado.iloc[i]['Supermercado'] == 'Disco':
    if 'pollo' in df_unificado.iloc[i]['Nombre Producto'].lower():
      df_unificado.iloc[i, df_unificado.columns.get_loc('Precio')] = (
                df_unificado.iloc[i]['Precio'] * 2.5
            )
```

```python
# Estándar de nombres:
# En base al nombre correspondiente se incorpora el nombre estandarizado del producto
# para futuras comparaciones

# Diccionario de elementos de Disco:
productos = ['Galletas de agua','Galletas dulces','Arroz','Harina','Papa',
             'Batata','Azucar','Cebolla','Manzana','Pollo',
             'Salame','Huevo', 'Queso crema', 'Yogur', 'Manteca',
             'Aceite','Gaseosa', 'Cerveza', 'Sal fina','Vinagre', 'Café',
             'Yerba', 'Mayonesa','Pan']

for i in range(len(df_unificado)):
    nombre_actual = df_unificado.iloc[i]['Nombre Producto'].lower()

    #  Condiciones específicas primero
    if 'galletas surtido' in nombre_actual or 'galletitas surtido' in nombre_actual:
        df_unificado.iloc[i, df_unificado.columns.get_loc('Nombre Producto')] = 'Galletas dulces'

    if 'galletitas sandwich mediatarde' in nombre_actual or 'galletitas de agua criollitas' in nombre_actual:
        df_unificado.iloc[i, df_unificado.columns.get_loc('Nombre Producto')] = 'Galletas de agua'

    if 'dulce de leche la serenisima clasico' in nombre_actual or 'dulce leche ilolay fliarx' in nombre_actual:
        df_unificado.iloc[i, df_unificado.columns.get_loc('Nombre Producto')] = 'Dulce de Leche'

    if 'leche entera la serenísima mult vit' in nombre_actual or 'leche entera la serenísima 3' in nombre_actual:
        df_unificado.iloc[i, df_unificado.columns.get_loc('Nombre Producto')] = 'Leche'

    # Reemplazos estándar del listado
    for producto in productos:
        if producto.lower() in nombre_actual:
            df_unificado.iloc[i, df_unificado.columns.get_loc('Nombre Producto')] = producto
```

## 4.  Análisis

#### a. Evolucion del precio promedio:

```python
# Evolucion del precio promedio, para observar como varia el precio promedio de los
# productos a lo largo del tiempo

# Agrupar por 'fecha'
evolucion = df_unificado.groupby('fecha')['Precio'].mean().reset_index()

# Ordenar  por fecha
evolucion = evolucion.sort_values('fecha')

# Grafico general
plt.figure(figsize=(10,5))
plt.plot(evolucion['fecha'], evolucion['Precio'], marker='o')
plt.title("Evolución General del precio promedio")
plt.xlabel("Fecha")
plt.ylabel("Precio promedio")
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

Se detecta una tendencia en baja del promedio de precios **del grupo de productos** analizados en la canasta basica. Dicha caida se produce entre el 17-04-2025 y el 19-04-2025.

```python
# Grafico de comparacion entre supermercado

# Disco:
disco_evolucion_general =  df_unificado[df_unificado['Supermercado'] == 'Disco']
disco_evolucion = disco_evolucion_general.groupby('fecha')['Precio'].mean().reset_index()

# Hiper Libertad:
hiper_libertad_evolucion_general =  df_unificado[df_unificado['Supermercado'] == 'Hiper Libertad']
hiper_evolucion = hiper_libertad_evolucion_general.groupby('fecha')['Precio'].mean().reset_index()

plt.figure(figsize=(10,5))
plt.plot(disco_evolucion['fecha'], disco_evolucion['Precio'], marker='o', label='Disco', color='red')
plt.plot(hiper_evolucion['fecha'], hiper_evolucion['Precio'], marker='o', label='Hiper Libertad', color='green')
plt.title("Comparacion de la Evolución del precio promedio por supermercado")
plt.xlabel("Fecha")
plt.ylabel("Precio promedio")
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

**Hiper Libertad** muestra una caída marcada en el precio promedio entre el 15 y el 19 de abril. Comienza con un precio promedio superior a \$2.900 y termina por debajo de \$2.350. La baja más significativa ocurre entre el 17-04-2025 y el 19-04-2025, donde se ve un descenso abrupto.

**Disco**, mantiene una tendencia estable con una leve disminución. Sus precios promedio oscilan cerca de \$2.550 durante todo el período.

Inicialmente, Hiper Libertad tenía precios notablemente más altos que Disco. Hacia el final del período, los precios de Hiper Libertad caen por debajo de Disco, invirtiendo la relación inicial.

Sin embargo, hay excepciones que podrían influir en la percepción general de precios.

* Disco no tiene precios publicados para: Café, Cebolla, Manzana, Pan.
* Hiper Libertad no tiene precios publicados para: Cerveza, Gaseosa, Mayonesa.

## c. Productos con mayor/menor variación de precios

```python
# Agrupar por producto y calcular medidas de dispersión
variacion = df_unificado.groupby('Nombre Producto')['Precio'].agg(['mean', 'std', 'min', 'max'])

# Calcular el rango de precios
variacion['rango'] = variacion['max'] - variacion['min']

# Calcular el coeficiente de variación
variacion['coef_var'] = variacion['std'] / variacion['mean']

# Ordenar productos por mayor variación (podés usar 'rango' o 'coef_var')
mayor_variacion = variacion.sort_values('coef_var', ascending=False)
menor_variacion = variacion.sort_values('coef_var', ascending=True)

# Ver los 10 productos mas y menos variables:
print(f'Top 10 Productos con mayor variación: \n {mayor_variacion.head(10)} \n')
print(f'\n Top 10 Productos con menor variación:\n  {menor_variacion.head(10)}\n')
```

Los productos con mayor variación de precios son principalmente alimentos frescos como pollo, salame, batata, y leche, que presentan altos valores en desviación estándar, rango y coeficiente de variación.
Los productos con menor variación presentan precios constantes, como café, cebolla y cerveza, todos con desviación estándar y rango igual a cero.

# 📌 Conclusión General del Análisis Exploratorio de Precios en Supermercados (Disco vs. Hiper Libertad)
El análisis desarrollado permitió observar la evolución de los precios de una canasta básica de productos en dos cadenas de supermercados de Córdoba (Argentina): Disco y Hiper Libertad, a lo largo de varios días consecutivos en abril de 2025. La integración, limpieza y estandarización de los datos fue un paso clave para asegurar la comparabilidad entre productos y formatos.

Entre los principales hallazgos, se identificó que **Hiper Libertad mostró una tendencia más pronunciada a la baja en el precio promedio de productos**, especialmente a partir del 17/04/2025. Esta caída podría estar asociada a estrategias de promociones, descuentos u optimización de surtido. En contraste, **Disco mantuvo precios más estables** a lo largo del período analizado, con variaciones menores.

Desde una perspectiva por producto, se observaron importantes diferencias entre supermercados. Algunos productos como el café, salame y pollo presentaron precios significativamente más altos, generando una gran dispersión. Por otro lado, productos como aceite, arroz y galletitas se mantuvieron con precios estables, lo cual sugiere una política de precios más uniforme en ciertos rubros.

A nivel de dispersión, los boxplots revelaron que Disco tiende a exhibir mayor variabilidad de precios, posiblemente asociada a una mayor diversidad de marcas y calidades ofrecidas, mientras que Hiper Libertad mostró precios más concentrados, aunque con ajustes marcados en los días finales del período observado.

Finalmente, las múltiples visualizaciones por producto ayudaron a detectar tendencias puntuales en alimentos esenciales, como la bajada de precios de lácteos, panificados y productos de almacén en Hiper Libertad, que en algunos casos terminaron con precios incluso menores que los de Disco, invirtiendo la relación inicial.
